```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('sales_data.csv')

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month_name()
df.head()


# Total Revenue
total_revenue = df['Amount'].sum()
print("Total Revenue:", total_revenue)

# Revenue by Service
service_revenue = df.groupby('Service')['Amount'].sum().sort_values(ascending=False)
print("\nRevenue by Service:")
print(service_revenue)

# Monthly Revenue
monthly_revenue = df.groupby('Month')['Amount'].sum()
print("\nMonthly Revenue:")
print(monthly_revenue)

# Plot Revenue by Service
service_revenue.plot(kind='bar')
plt.title('Revenue by Service')
plt.show()
```

**ðŸ” Key Insights**

* AMC Contracts generate the highest revenue among all services.

* Lawn Development is the most frequent service.

* Referral leads bring higher-value projects compared to social media.

* Revenue shows consistent demand across months.
